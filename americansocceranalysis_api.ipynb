```python
endpoint_url = "https://app.americansocceranalysis.com/api/v1/mls/players"
```

```python
response = requests.get(endpoint_url)
response.status_code
```

```python
print(response.json())
```

```python
first_player_name = response.json[0]['player_name']
print("First player name:", first_player_name)
```

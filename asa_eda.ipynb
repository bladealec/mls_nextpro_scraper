```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mlsnp_players_df = pd.read_csv('data/MLSNP_players.csv')
mlsnp_teams_df = pd.read_csv('data/MLSNP_teams.csv')
mlsnp_games_df = pd.read_csv('data/MLSNP_games.csv')
```

```python
print("Basic Info - MLS Next Pro Players:")
print(mlsnp_players_df.info())
```

```python
print("\nSummary Statistics - MLS Next Pro Players:")
print(mlsnp_players_df.describe())
```

```python
# distribution of player heights
plt.figure(figsize=(10, 6))
sns.histplot(mlsnp_players_df["height_ft"], bins=20, kde=True, color='blue')
plt.title("Distribution of Player Heights (MLS Next Pro)")
plt.xlabel("Height (feet)")
plt.ylabel("Frequency")
plt.show()
```

```python
# distribution of player weights
plt.figure(figsize=(10, 6))
sns.histplot(mlsnp_players_df["weight_lb"], bins=20, kde=True, color='green')
plt.title("Distribution of Player Weights (MLS Next Pro)")
plt.xlabel("Weight (lbs)")
plt.ylabel("Frequency")
plt.show()
```
